```python
import pandas as pd
```

```python
df = pd.read_csv("BeerDataScienceProject.tar.bz2", compression='bz2', sep=',')
df.head()
```

1. Rank the top 3 breweries which produce the strongest beers.

```python
breweries_producing_strongestbeer_df=df.groupby("beer_brewerId")["beer_ABV"].mean().sort_values(ascending=False).head(3).reset_index()
breweries_producing_strongestbeer_df
```

2. Which year did beers enjoy the highest ratings?

```python
df['review_year']=pd.to_datetime(df['review_time']).dt.year
```

```python
year_with_highest_rating=df.groupby('review_year')["review_overall"].mean().sort_values(ascending=False)
year_with_highest_rating
```

3. Based on the users' ratings, which factors are important among taste, aroma, appearance, and palette?

```python
important_factors_df=df.groupby('beer_beerId')[["review_taste","review_aroma","review_appearance","review_palette","review_overall"]].mean()
important_factors_df.corr()
```

From the above correlation between all review columns, "aroma", "palettee" and then "taste" is very important respectively for users as compared to nay other factor.

4. If you were to recommend 3 beers to your friends based on this data, which ones would you recommend?

```python
beers_df=df.groupby("beer_name")[["review_aroma","review_appearance","review_palette"]].mean().sort_values(by=["review_aroma","review_appearance","review_palette"],ascending=False)
beers_df.head(3)
```

5. Which beer style seems to be the favourite based on the reviews written by users? How does written reviews compare to overall review score for the beer style?

```python
favourite_beer_df=df.groupby(by="beer_style")["review_overall"].mean().sort_values(ascending=False).head(1).reset_index()
favourite_beer_df
```
